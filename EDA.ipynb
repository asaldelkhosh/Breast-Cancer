```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dc = load_breast_cancer(as_frame=True)
dc = dc.frame
```

```python
print("\nNull Values:\n", dc.isnull().sum())
print("\nMissing Values:\n", dc.isna().sum())
```

```python
dc.info()
```

```python
dc.describe()
```

```python
dc['target'].value_counts()
```

```python
dc['target'].value_counts() \
    .plot(kind="bar", width=0.1, color=["lightgreen", "cornflowerblue"], legend=1, figsize=(8, 5))
plt.xlabel("(1 = Benign) (0 = Malignant)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(fontsize=12);
plt.yticks(fontsize=12)
plt.legend(["Benign"], fontsize=12)
plt.show()
```

```python
plt.figure(figsize =(30,30))
sns.heatmap(dc.corr(), vmin=-1, vmax=1, annot=True)
plt.show()
```
